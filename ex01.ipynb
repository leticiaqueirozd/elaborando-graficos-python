```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'imdb_movie_dataset.csv'
imdb_data = pd.read_csv(file_path)
```

```python
# barras
top_movies_revenue = imdb_data.nlargest(10, 'Revenue (Millions)')
titles = top_movies_revenue['Title']
revenues = top_movies_revenue['Revenue (Millions)']

plt.figure(figsize=(10, 6))
plt.bar(titles, revenues, color='pink')
plt.title('Top 10 Filmes por $$$')
plt.xlabel('Título do Filme')
plt.ylabel('Receita em milhões')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# pizza
genres = imdb_data['Genre'].str.split(',', expand=True).stack().value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(genres, labels=genres.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Gêneros de filmes')
plt.show()
```

```python
# dispersão

plt.figure(figsize=(10, 6))
plt.scatter(imdb_data['Votes'], imdb_data['Revenue (Millions)'], alpha=0.7, color='purple')
plt.title('Receita X Votos')
plt.xlabel('Número de Votos')
plt.ylabel('Receita em milhões')
plt.grid(True)
plt.show()
```
